## Data CLeaning

```python
#Replacing missing value with NaN
df.replace(['-', 'N.A', 'N,A'], pd.NA, inplace=True)
```

```python
# Converting numerical columns to proper numeric types
numeric_cols = ['Goals Scored ', 'Assists Provided ', 'Dribbles per 90', 
                'Interceptions per 90', 'Tackles per 90', 'Total Duels Won per 90',
                'Save Percentage', 'Clean Sheets']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

```

```python
# Handling missing values (here I filled with 0 for performance metric)
performance_cols = ['Goals Scored ', 'Assists Provided ', 'Dribbles per 90', 
                   'Interceptions per 90', 'Tackles per 90', 'Total Duels Won per 90']
df[performance_cols] = df[performance_cols].fillna(0)
```

# For goalkeepers
gk_mask = df['Position'] == 'GK'
df.loc[gk_mask, 'Save Percentage'] = df.loc[gk_mask, 'Save Percentage'].fillna(
    df.loc[gk_mask, 'Save Percentage'].median())

## Exploratory Data Analysis (EDA)

### Basic Statistics

```python
# Summary statistics for numerical columns
print(df.describe())

# Count of players by position
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Position', order=df['Position'].value_counts().index)
plt.title('Number of Players by Position')
plt.xticks(rotation=45)
plt.show()
```

### Top Performers Analysis

```python
# Top 10 goal scorers
top_scorers = df.nlargest(10, 'Goals Scored ')[['Player Name ', 'Nationality ', 'Position', 'Goals Scored ']]
print("Top 10 Goal Scorers:") 
print(top_scorers)
```

```python

# Top 10 assist providers
top_assists = df.nlargest(10, 'Assists Provided ')[['Player Name ', 'Nationality ', 'Position', 'Assists Provided ']]
print("\nTop 10 Assist Providers:")
print(top_assists)
```

### NAtional Teams Analysis

```python
# Goals by country
goals_by_country = df.groupby('Nationality ')['Goals Scored '].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
goals_by_country.plot(kind='bar')
plt.title('Top 10 Countries by Total Goals Scored')
plt.ylabel('Total Goals')
plt.show()

```

```python
# Average goals per player by country (with at least 10 players)
country_stats = df.groupby('Nationality ').agg(
    Player_Count=('Player Name ', 'count'),
    Total_Goals=('Goals Scored ', 'sum'),
    Avg_Goals=('Goals Scored ', 'mean')
).query('Player_Count >= 10').sort_values('Avg_Goals', ascending=False).head(10)

print("\nCountries with highest average goals per player (min 10 players):")
print(country_stats)
```

### Brand Sponsorship Analysis

```python
# Most common brand sponsors
brand_counts = df['Brand Sponsor/Brand Used'].value_counts().head(10)
plt.figure(figsize=(10,6))
brand_counts.plot(kind='bar')
plt.title('Top 10 Player Sponsorships')
plt.ylabel('Number of Players')
plt.show()
```

```python

# Performance by brand (for brands with at least 20 players)
brand_stats = df.groupby('Brand Sponsor/Brand Used').agg(
    Player_Count=('Player Name ', 'count'),
    Avg_Goals=('Goals Scored ', 'mean'),
    Avg_Assists=('Assists Provided ', 'mean')
).query('Player_Count >= 20').sort_values('Avg_Goals', ascending=False)

print("\nPerformance Metrics by Brand (min 20 players):")
print(brand_stats)
```

## Visualization

```python
# Correlation heatmap
plt.figure(figsize=(12,8))
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Performance Metrics')
plt.show()
```

```python

# Goals vs Assists by Position
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='Goals Scored ', y='Assists Provided ', hue='Position')
plt.title('Goals vs Assists by Position')
plt.show()

```

```python
# Top 10 nations by goals with average per player
top_countries = df.groupby('Nationality ')['Goals Scored '].sum().nlargest(10).index
country_goals = df[df['Nationality '].isin(top_countries)]

plt.figure(figsize=(12,6))
sns.boxplot(data=country_goals, x='Nationality ', y='Goals Scored ')
plt.title('Goals Distribution for Top 10 Scoring Nations')
plt.xticks(rotation=45)
plt.show()
```
